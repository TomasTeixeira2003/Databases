## 1 Conversão E-A para Modelo Relacional

Recorde a notação para o Modelo Relacional exposta nas aulas teóricas como se segue:

- relacaoA (<u>atributo1</u>, … , atributoN)
    - atributoX, … , atributoZ: FK (RelacaoDestino)
    - unique(atributoA, … , AtributoC)
- relacaoB (<u>atributo1</u>, … , atributoN)
    - atributoX, … , atributoZ: FK (RelacaoDestino)
    - unique(atributoA, … , AtributoC)

1. As relações são expressas por um nome seguido dos atributos entre parêntesis;
    - As chave primárias são expressas sublinhando os atributos que a compõem;
    - As chaves estrangeiras para outras relações são expressas através da lista de atributos que a compõem, seguida de “: FK(RelacaoDestino)”. A ordem dos atributos que compõem a FK deve corresponder à ordem dos atributos da chave na relação destino;
    - As chaves candidatas são expressas através da expressão “unique”, seguida da lista de atributos que a compõem entre parêntesis;
    - As restrições de integridade ao Modelo Relacional são descritas em termos dos conceitos deste modelo.

- IC-1 : When an Order_ is removed from the database it must also be removed from Sale
- IC-2: any sku in Product must exist in Supplier
- IC-3 : any order_no in Order_ must exist in contains
- IC-4 : when a Product is removed from the database it must also be removed from EAN Product
- IC-5: any ssn in Employee must exist in works
- IC-6: when a Workplace is removed from the database it must also be removed from Office and/or Warehouse

### 3.1 Esquema Relacional (schema.sql)

Crie uma base de dados SQL correspondente ao esquema Relacional desenvolvido no ponto anterior.
Utilizando a linguagem SQL (DDL), apresente as instruções para criar o esquema de bases de dados correspondente ao modelo relacional desenvolvido. Deve certificar-se de que os tipos de dados e tamanhos de campo selecionados sejam os mais apropriados. As restrições aplicáveis a cada campo, linha e tabela também devem ser especificadas recorrendo às instruções NOT NULL, CHECK, PRIMARY KEY, UNIQUE e FOREIGN KEY conforme seja apropriado.

```python
%load_ext sql
%sql postgresql://db:db@postgres/db
```

```python
%%sql

drop table if exists Customer cascade;
drop table if exists Order_ cascade;
drop table if exists Sale cascade;
drop table if exists pay cascade;
drop table if exists Product cascade;
drop table if exists contains cascade;
drop table if exists EAN_Product cascade;
drop table if exists Employee cascade;
drop table if exists process cascade;
drop table if exists Supplier cascade;
drop table if exists Department cascade;
drop table if exists Workplace cascade;
drop table if exists works cascade;
drop table if exists Office cascade;
drop table if exists Warehouse cascade;
drop table if exists delivery cascade;
-- schema.sql

CREATE TABLE Customer (
    cust_no NUMERIC,
    name VARCHAR,
    email VARCHAR UNIQUE,
    phone VARCHAR(9),
    address VARCHAR,
    PRIMARY KEY (cust_no));

CREATE TABLE Order_ (
    order_no NUMERIC,
    date_ DATE,
    cust_no NUMERIC NOT NULL,
    PRIMARY KEY (order_no),
    FOREIGN KEY (cust_no) REFERENCES Customer(cust_no) ON DELETE CASCADE);

CREATE TABLE Sale (
    order_no NUMERIC,
    PRIMARY KEY (order_no),
    FOREIGN KEY (order_no) REFERENCES Order_(order_no) ON DELETE CASCADE);

CREATE TABLE pay (
    order_no NUMERIC,
    cust_no NUMERIC NOT NULL,
    FOREIGN KEY (cust_no) REFERENCES Customer(cust_no),
    FOREIGN KEY (order_no) REFERENCES Order_(order_no),
    PRIMARY KEY (order_no));

CREATE TABLE Product (
    sku NUMERIC(8),
    name VARCHAR,
    description VARCHAR,
    price DECIMAL(10, 2) CHECK (price > 0),
    PRIMARY KEY (sku));
    /* IC-1: any sku in Product must exist in Supplier */

CREATE TABLE contains (
    order_no NUMERIC NOT NULL,
    sku NUMERIC(8),
    qty NUMERIC CHECK (qty > 0),
    PRIMARY KEY (order_no, sku),
    FOREIGN KEY (order_no) REFERENCES Order_(order_no),
    FOREIGN KEY (sku) REFERENCES Product(sku));
    /* IC-2 : any order_no in Order must exist in contains */


CREATE TABLE EAN_Product (
    sku NUMERIC(8),
    ean VARCHAR(13),
    PRIMARY KEY (sku),
    FOREIGN KEY (sku) REFERENCES Product(sku) ON DELETE CASCADE);

CREATE TABLE Employee (
    ssn VARCHAR(11),
    PRIMARY KEY (ssn),
    TIN NUMERIC(9) UNIQUE,
    bdate DATE,
    name VARCHAR);

CREATE TABLE process (
    order_no NUMERIC,
    ssn VARCHAR(11),
    PRIMARY KEY (order_no, ssn),
    FOREIGN KEY (order_no) REFERENCES Order_(order_no),
    FOREIGN KEY (ssn) REFERENCES Employee(ssn));

CREATE TABLE Supplier (
    TIN NUMERIC(9),
    PRIMARY KEY (TIN),
    name VARCHAR,
    address VARCHAR,
    sku NUMERIC(8) NOT NULL,
    date_ DATE,
    UNIQUE (sku, TIN),
    FOREIGN KEY (sku) REFERENCES Product(sku));

CREATE TABLE Department (
    name VARCHAR,
    PRIMARY KEY (name));

CREATE TABLE Workplace (
    address VARCHAR,
    PRIMARY KEY (address),
    lat DECIMAL(9, 6),
    long_ DECIMAL(9, 6),
    UNIQUE (lat, long_));

CREATE TABLE works (
    ssn VARCHAR(11) NOT NULL,
    name VARCHAR,
    address VARCHAR,
    PRIMARY KEY (ssn, name, address),
    FOREIGN KEY (ssn) REFERENCES Employee(ssn),
    FOREIGN KEY (name) REFERENCES Department(name),
    FOREIGN KEY (address) REFERENCES Workplace(address));
    /* IC-3: any ssn in Employee must exist in works */


CREATE TABLE Office (
    address VARCHAR,
    PRIMARY KEY (address),
    FOREIGN KEY (address) REFERENCES Workplace(address) ON DELETE CASCADE);

CREATE TABLE Warehouse (
    address VARCHAR,
    PRIMARY KEY (address),
    FOREIGN KEY (address) REFERENCES Workplace(address) ON DELETE CASCADE);

CREATE TABLE delivery (
    address VARCHAR,
    sku NUMERIC(8),
    TIN NUMERIC(9),
    PRIMARY KEY (address, sku, TIN),
    FOREIGN KEY (address) REFERENCES Warehouse(address),
    FOREIGN KEY (sku, TIN) REFERENCES Supplier(sku, TIN));
```

## 4 Linguagem SQL

Apresente a expressão SQL correspondente  a cada uma das interrogações de Álgebra Relacional.


1. Liste o nome de todos os clientes que fizeram encomendas contendo produtos de preço superior a €50 no ano de 2023;

```python
%%sql

SELECT DISTINCT c.name
FROM Customer c
JOIN Order_ o USING (cust_no)
JOIN contains ct USING (order_no)
JOIN Product p USING (sku)
WHERE p.price > 50 AND EXTRACT(YEAR FROM o.date_) = 2023;
```

2. Liste o nome de todos os empregados que trabalham em armazéns e não em escritórios e processaram encomendas em Janeiro de 2023;

```python
%%sql

SELECT e.name
FROM Employee e
JOIN process pr USING (ssn)
JOIN Order_  o USING (order_no)
JOIN works USING (ssn)
JOIN Workplace w USING (address)
JOIN Warehouse wa USING (address)
WHERE w.address = wa.address
    AND EXTRACT(YEAR FROM o.date_) = 2023
    AND EXTRACT(MONTH FROM o.date_) = 1
EXCEPT
SELECT e.name
FROM Employee e
JOIN works USING (ssn)
JOIN Workplace w USING (address)
JOIN Office off  USING (address)
WHERE w.address = off.address;
```

3. Indique o nome do produto mais vendido;

```python
%%sql
--if there is more than one most sold then all of them appear


SELECT name AS most_sold_product
FROM
(SELECT name, SUM(qty) AS quantity 
    FROM (Product p 
    JOIN contains c USING (sku)) 
    GROUP BY name) AS q1
JOIN
    (SELECT MAX(quantity) AS quantity
    FROM
    (SELECT name, SUM(qty) AS quantity 
        FROM (Product p 
        JOIN contains c USING (sku)) 
        GROUP BY name) AS q2) AS q3 USING(quantity);
```

4. Indique o valor total de cada venda realizada.

```python
%%sql
-- Order 1001 exists but was not paid, therefore it does not appear
SELECT s.order_no, SUM(p.price * ct.qty) AS total_value
FROM Sale s
JOIN Contains ct USING (order_no)
JOIN Product p USING(sku)
GROUP BY s.order_no;

```
